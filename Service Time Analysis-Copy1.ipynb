```python
import seaborn as sns
#process based on date first

def getLabel(s):
    a=int(s.split(' ')[0].split('-')[-1])
    if(a==9):
        return('First Day')
    else:
        return('Last Day')

    
df['label'] = df['last_started_at'].apply(getLabel)
```

```python
print(df.head())
print(df.tail())
```

```python
sns.barplot(x='service',y='boot_time_in_sec',hue='label',data=df)
```
