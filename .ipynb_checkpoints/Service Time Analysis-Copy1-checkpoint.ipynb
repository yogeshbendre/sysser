```python
%matplotlib inline
df.index = df["Service"]
df.plot.bar()
```

```python
import seaborn as sns
#process based on date first

def getLabel(s):
    a=int(s.split(' ')[0].split('-')[-1])
    if(a==9):
        return('First Day')
    else:
        return('Last Day')

    
df['label'] = df['LastBootedAt'].apply(getLabel)
```

```python
print(df.head())
print(df.tail())
```

```python
sns.barplot(x='Service',y='BootTimeInSec',hue='label',data=df)
```

```python
dfLast = df[df['label']=='Last Day']
```

```python
dfLast
```

```python
dfLast = dfLast.sort_values(by='BootTimeInSec',ascending=False)
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
plt.figure(figsize=(30,15))
g = sns.barplot(data=dfLast,x='Service',y='BootTimeInSec')
#g.set_xticklabels(rotation=0)
plt.xticks(rotation=90,fontsize=18)
plt.xlabel('Service',fontsize=18)
plt.yticks(fontsize=18)
plt.ylabel('Boot Time In Sec',fontsize=18)
```

```python
plt.figure(figsize=(20,10))
sns.distplot(dfLast['BootTimeInSec'],kde=False,rug=True,bins=24)
plt.xticks(fontsize=18)
plt.yticks(fontsize=18)
plt.xlabel('Boot Time In Sec',fontsize=18)
plt.ylabel('Number of Services',fontsize=18)
```

```python
plt.figure(figsize=(20,10))
sns.distplot(dfLast['BootTimeInSec'],kde=False,rug=True,bins=24,hist_kws=dict(cumulative=True))
plt.xticks(fontsize=18)
plt.yticks(fontsize=18)
plt.xlabel('Time In Sec',fontsize=18)
plt.ylabel('Number of Services Started',fontsize=18)
```

```python
import pandas as pd
df=pd.read_csv("Cloud02_22_05_2019.txt",sep="|")
#print(df)
df['date'] = df['LastBootedAt'].apply(lambda x : x.split(' ')[0])
df2 = df.loc[:,['Service','BootTimeInSec','date']]
print(df2.head())
```

```python
df3=df2.pivot_table(columns='date',index='Service',fill_value=-1)
```

```python
%matplotlib inline

sns.heatmap(data=df3)
```

```python
#sns.set_context('talk')
df4 = df2[df2['date']!='2019-05-17']
print(df4.head())
plt.figure(figsize=(30,15))
sns.barplot(data=df4,x='Service',y='BootTimeInSec',hue='date')
plt.xticks(rotation=90)
```

```python
df22 = df4[df4['date']=='2019-05-22']
df22=df22.sort_values(by='BootTimeInSec',ascending = False)
df22.head()
```

```python
df21 = df4[df4['date']=='2019-05-21']
df21=df21.sort_values(by='BootTimeInSec',ascending = False)
df21.head()
```
