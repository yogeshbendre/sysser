```python
import pandas as pd
df=pd.read_csv("LDU40.txt",sep="|")
print(df)
```

```python
#df=df.drop(["LastBootedAt"],axis=1)
```

```python
%matplotlib inline
df.index = df["Service"]
df.plot.bar()
```

```python
import seaborn as sns
#process based on date first

def getLabel(s):
    a=int(s.split(' ')[0].split('-')[-1])
    if(a==9):
        return('First Day')
    else:
        return('Last Day')

    
df['label'] = df['LastBootedAt'].apply(getLabel)
```

```python
print(df.head())
print(df.tail())
```

```python
sns.barplot(x='Service',y='BootTimeInSec',hue='label',data=df)
```

```python
dfLast = df[df['label']=='Last Day']
```

```python
dfLast
```

```python
dfLast = dfLast.sort_values(by='BootTimeInSec',ascending=False)
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
plt.figure(figsize=(30,15))
g = sns.barplot(data=dfLast,x='Service',y='BootTimeInSec')
#g.set_xticklabels(rotation=0)
plt.xticks(rotation=90,fontsize=18)
plt.xlabel('Service',fontsize=18)
plt.yticks(fontsize=18)
plt.ylabel('Boot Time In Sec',fontsize=18)
```

```python
plt.figure(figsize=(20,10))
sns.distplot(dfLast['BootTimeInSec'],kde=False,rug=True,bins=24)
plt.xticks(fontsize=18)
plt.yticks(fontsize=18)
plt.xlabel('Boot Time In Sec',fontsize=18)
plt.ylabel('Number of Services',fontsize=18)
```

```python
plt.figure(figsize=(20,10))
sns.distplot(dfLast['BootTimeInSec'],kde=False,rug=True,bins=24,hist_kws=dict(cumulative=True))
plt.xticks(fontsize=18)
plt.yticks(fontsize=18)
plt.xlabel('Time In Sec',fontsize=18)
plt.ylabel('Number of Services Started',fontsize=18)
```
